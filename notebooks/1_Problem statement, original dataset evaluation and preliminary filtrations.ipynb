## Introduction

Spain has been characterized as having a great predisposition to experience strong fires, specially during the summer season (a fact experienced by most regions with a mediterranean climate). The impact of these fires extends beyond the obvious ecological catastrophe that they represent at a local level, local communities are aslo affected, provoking, in many ocassions, an irreparable damage to private and public property. At an economic level, forest fires have a great impact in some of the main sectors of these rural communities, extending the damage beyonde the geographic or ecological frame. 

## Original dataset: source,  intended use and data description

***1.1 Data source***

As we mentioned before, fire infromation is derived by the official reports fulfilled by the competent authorities in matters of fire extinction (a description of the data can be found in the official filling instructions in the following [link](https://www.miteco.gob.es/es/biodiversidad/temas/incendios-forestales/instrucciones_parte_incendio_tcm30-512355.pdf)), the historical data is already available thanks to the effort of **[CIV10](https://civio.es/nosotros/)** through their proyect **[España en llamas](https://civio.es/espana-en-llamas/metodologia/)** in which a web application allows viewing and browsing unpublished data of all forest fires with an affected surface of more than one hectare occurred in Spain between 2001 and 2015 (more than 80,000 observations).

***1.2 Intended use***

The intention of the present work is the **generation of a binary classification model that can help to predict if the causes that produced a fire were accidental or intentional in the Iberian Peninsula territory**. In order to achieve this task, documentation from a diverse range of sources will be gradually added and/or combine (fire, climate data and geographic information), so data quality could be evaluated in different stages of the analysis and through different machine learning techniques and deep learning algorithms.

* The first step, and the one that involves the use of the data collected by **[CIV10](https://civio.es/nosotros/)** , will be the modeling of this information (before adding the rest of variables) in order to evaluate the sufficiency of the data provide by the official fire reports to achieve the task of predicting the cause of a fire. (A second phase will be performe in order to evaluate whether or not climate and geographic information should be added as variables into the model, considering a previous analysis of the effects of these atributes over our predicted variable through exploratory data analysis and, in case of addition, how this atributes may result in a better or worst performance of the models).


***1.3 Data description***

The dataset has 82,640 observations taken during the years 2001 and 2015 (last years with **complete** information available) and contains the following fields:

- **ID**: Fire identifier
- **superficie**: Affected surface
- **fecha**: Fire date (format yyyy-mm-dd)
- **lat**: 	Geographic latitude of fire origine (WGS84)
- **lng**: Geographic longitude of fire origine (WGS84)
- **latlng_explicit**: If fire coordinates have been used (1) if coordinates of the closest municipality has been used (0)
- **idcomunidad**: Autonomous community identifier. (range 1-18)
- **idprovincia**: Province identifier (range 1-51)
- **idmunicipio**: Municipality identifier (codes vary from 1 to 999 in a none consecutive order)
- **municipio**: Municipality name
- **causa**: Cause identifier, with the following subclasification:


|cause|Code|
|:---:|:-------------:|
|fire caused by lightning|1|
|accident or negligence|2|
|accident or negligence|3|
|intentional fire|4|
|unknown cause|5|
|reproduced fire|6|


- **causa_supuesta**: (1) for supposed causes (causes not totally sure or completly verify), in case of complete certainty none or nan value
- **causa_desc**: Identifier wit a more complex description of the causes.
- **muertos**: 	Number of dead people due to the fire.
- **heridos**: Number of harmed people due to the fire.
- **time_ctrl**: Time to control the fire (in minutes)
- **time_ext**: Time to completly extinguished the fire (in minutes)
- **personal**: Number of people involved in the firefighting activities.
- **medios**: Number of logistic and technique resources that have participated in the extinction of the fire
- **gastos**: expense
- **perdidas**: economic loss

## Loading data and subsetting by the Iberian Peninsula of the spanish territory

### Loading data 

```python
data_path = '../Data'
all_fires_path = 'incendios_2001_215.csv'
codes = 'codes.csv'
all_fires_df = pd.read_csv(os.path.join(data_path, all_fires_path))
codes_df = pd.read_csv(os.path.join(data_path, codes), sep = ';', encoding='latin-1')
geodata_path = 'geo_data'
poligons_path = 'recintos_provinciales_inspire_peninbal_etrs89.zip'
df_poligons = gpd.read_file(os.path.join(data_path, geodata_path, poligons_path))
```

### Subsetting

```python
#Explore provinces contained in df_poligons
df_poligons.NAMEUNIT.unique()
```

**Spain has several territories overseas such as *Islas Canarias* e *Islas Baleares* and also cities in the north of Africa like *Ceuta* and *Melilla*, these territories are not part of the Peninsula Iberica, so they wont be consider for these analysis.**

```python
#coordinate proyection system from shapefile:
df_poligons.crs
```

```python
#Subsetting Peninsular data only, crossing 
exclusion = ['Illes Balears','Ceuta', 'Melilla', 'Territorio no asociado a ninguna provincia']
df_peninsula = df_poligons[~df_poligons['NAMEUNIT'].isin(exclusion)]
#Selecting the coordinate proyection for crossing data (it has to be the same proyection system)
coordinates_sys = 'EPSG:4258'
#Creation of geodataframe
#We create a Point entinty for each fire observation
geometry = [Point(xy) for xy in zip(all_fires_df.lng, all_fires_df.lat)]
#create a geodf from all_fires_df, we name the colum of all those points geometry and assign the correct crs proyection
geo_incendio_df = gpd.GeoDataFrame(all_fires_df, geometry = geometry, crs = coordinates_sys)
#We cross data to create a unique dataframe with peninsular data only
all_fires_df = gpd.sjoin(geo_incendio_df, df_peninsula[['geometry']], op='within', how='inner')
```

## Missing data evaluation

```python
#Evaluaation of missing data and datatypes
all_fires_df.info()
```

```python
#proportion of missing data per columns:
total_obs = all_fires_df.shape[0]
print('Percentage of missing values of atributes with more than a 0% of missing values:')
for variable in all_fires_df.columns:
    missing = all_fires_df[variable].isnull().sum().sum()
    percentage = int((missing/total_obs)*100)#redondeamos a entero para evitar la incrusión de datos con más de un 99% de datos faltantes
    if percentage != 0:
        print('{}: {}%'.format(variable, percentage))
```

##  Restructuring dataframe: changing date datatype, adding month and year as an atribute

```python
#changing format of date atribute into a datatype datetime:
all_fires_df['fecha'] = pd.to_datetime(all_fires_df['fecha'])
#Adding year:
all_fires_df['year'] = all_fires_df['fecha'].map(lambda date: int(date.strftime('%Y')))
#Adding month:
all_fires_df['month'] = all_fires_df['fecha'].map(lambda date: int(date.strftime('%m')))
#ordering the dataframe by date:
all_fires_df.sort_values(by = 'fecha', inplace=True)
```

### Restructuring dataframe: evaluation per trimester

```python
#create a fnction that calculates the proportion or the ocurrencies of certain class of a column in the data
def cause_per_class(df, field, count = False):
    if count:
        return df[field].value_counts().sort_index()
    return (df[field].value_counts().sort_index() / df.shape[0])*100
```

```python
all_fires_df['class_per_trimester'] = pd.cut(x=all_fires_df['month'], 
                                        bins=[1, 3, 6, 9, 12],
                                        labels = ['1', '2', '3','4'])
```

## Observations, extracting preliminary insights from EDA

In the sample, we observe correlations of all kinds. Variables that we might intuitively think are more correlated, such as fire size and extinguishing time (larger fire size leading to longer required control or extinguishing time), express moderate relationships. On the other hand, the resources used in extinguishing activities, both personnel and equipment, show relatively low levels of correlation with the surface variable.

It's possible that some correlations, with high indices like the one found between equipment and personnel, indicate attribute redundancy. This means that we might have variables that are somehow describing the same thing (high collinearity). This makes sense in the example of equipment and personnel; these two variables represent the resources used in extinguishing activities, whether they are human, technical, or logistical resources. Considering using only one of these variables or some combination of them for building the predictive model would be reasonable to avoid biasing the prediction.

Within this context, we could also consider the correlation between expenses and 'resources' (whether equipment or personnel). Expenses could be an indirect measure of the resources used for fire extinguishing. The remaining correlations (with slightly more extreme values) might be due to the lack of variables contributing to the analysis, such as climatic or spatial variables (maximum temperatures, wind speed, radiation, relative humidity, topography, fuel material, soil type, etc.), which play a tremendously important role in fire development.

Another important piece of information is the significant imbalance between intentional and accidental fires in the sample. Please refer to the figure generated in section 2.7, where we can observe that roughly half of the causes are intentional. Initially, this doesn't pose a problem due to the transformations and filtrations yet to be performed (since all non-intentional causes will be grouped into a single class for the binary classification model). However, we should consider that these circumstances might repeat in the final dataset and act accordingly. A large imbalance between evaluated classes could bias the model's prediction toward the cause with the higher number of records.

In the context of fire-cause, intentional fires are the primary cause of fires regardless of data segmentation or time scale, such as fire duration, cause-month relationship, or causes per quarter (figures 2.14 and 2.15).

Regarding the distributions of surface area, extinguishing time, and control time per year, it's worth noting the significant differences in magnitude between central measures and outlier values (outliers), figures generated in sections 2.5 and 2.7. Decisions will have to be made during data processing to manipulate them correctly without biasing the model.

## Geografical data, pipeline for preliminary filtration:

- **Regarding Geolocation**: The **lat** and **lng** data will be crucial to adapt the additional information that will be added later, especially considering that a significant portion of that information will come from geographical data. All observations not specifying location based on these two columns will be removed from the dataset.

- For enhanced model accuracy, we will filter the dataset considering only the **1** values in the **latlng_explicit** field, as they reflect the true position of the fire and thus the real values of the data that we will incorporate later on.

- The **idcomunidad** and **idmunicipio** data will initially be used to visualize the distribution and evolution of fires over time, but they are variables contained within the geographical coordinates and will not be used for analysis.

- For the purpose of this exercise, only data from the **Iberian Peninsula** will be considered for analysis.

```python
#Elimination of rows without geo info
all_fires_df_lat_lng = all_fires_df.dropna(subset=['lat','lng'])
# filtration by no explicit coordinates
all_fires_df_lat_lng_1 = all_fires_df_lat_lng[all_fires_df_lat_lng['latlng_explicit'] == 1]
```

## Process and filtration by cause, creating the Y to be predicted by our model:

- **Regarding Causes**: As previously mentioned, the purpose of this work is to develop a binary classification model to determine whether a fire is intentional or not. To achieve this, we will need to modify the **cause** column by grouping all categories related to accidents, negligence, or natural causes into a single group. Another category will encompass solely intentional fires.

   In the case of fires labeled as unknown (code ), they will be extracted from the training dataset and used to predict their most likely category using the final model.
    
   Observations evaluated as **1** (*presumed*) in the **causa_supuesta** field will also be removed from the training set, as they are not considered 100% certain causes indicated in the reports. To contrast the model's decisions with expert investigation, we will evaluate these data using the final model.
    
   Regarding reproduced fires, according to the [regulations](https://www.miteco.gob.es/es/biodiversidad/temas/incendios-forestales/instrucciones_parte_incendio_tcm30-512355.pdf), this cause should not be indicated in the case of intentional fires that are repeated in the same location or when the cause of reproduction is external to the fire itself. Therefore, they will be added to the category of unintentional fires. The model distinguishes between intentional and accidental causes, so we will generate a 'binary_causes' column where intentional causes will take a value of 1, and all other causes will be grouped into a single value, 0.

```python
all_fires_df_causes = all_fires_df_lat_lng_1[(all_fires_df_lat_lng_1['causa'] != 5)&(all_fires_df_lat_lng_1['causa_supuesta'] !=1)]
all_fires_df_causes['Y'] = all_fires_df_causes['causa'].apply(lambda x: 1 if x == 4 else 0)
display(cause_per_class(all_fires_df_causes, 'Y'))
```

**As reflected in the cell above, after creating the response variable $Y$, the sample becomes much more balanced by considering the grouping of unintentional causes into a single category.**

## Some last decisions, dealing with final nans:

**The rest of the fields with over 50% missing data**:

As we saw previously in section 4, the proportion of missing data ranged between 70% and 90% for some of the above-mentioned features prior to filtration:

|Field|Percentage of Missing Data|
|:---:|:-----------------------:|
|muertos (deceased) |96|
|heridos (injured) |96|
|gastos (expenses) |85|
|perdidas (losses) |71|

In fields like deceased or injured, missing data refers to zero deceased or zero injured individuals. Considering that the number of affected individuals in this context doesn't contribute to the analysis due to the majority of incidents having no affected individuals, the same value is repeated across almost all observations. We will discard these attributes for future analysis.

Regarding the expenses variable, it will be excluded from the analysis due to its approximately 90% missing data and the potential redundancy it might introduce to the model, given its moderately high correlation with the variables equipment and personnel (human, technical, and logistical resources used in extinguishing activities).

```python
#Proportion of missing data after filtration
total_obs = all_fires_df_causes.shape[0]
fields = []
print('Columns with missing values and its percentages:\n')
for variable in all_fires_df_causes.columns:
    missing = all_fires_df_causes[variable].isnull().sum().sum()
    percentage = int((missing/total_obs)*100)
    if percentage != 0:
        print('{}: {}% of missing data'.format(variable, percentage))
        if percentage > 50:
            fields.append(variable)
```

After the filtration based on presumed causes, the column "causa_supuesta" only contains NaN values as a result of the filtration carried out in previous sections Therefore, it will be removed from the dataset.

```python
for variable in fields:
    all_fires_df_causes.drop(variable, inplace = True, axis = 1)
```

Regarding resources, both the variable "personal" and "gastos" indirectly refer to the same attribute, even though they refer to different resources (i.e., the "personal" column doesn't encompass the same resources as the "equipment" column). We will group these two columns under a single attribute, named recursos (resources).

```python
all_fires_df_causes['recursos'] = all_fires_df_causes['personal'] + all_fires_df_causes['medios']
all_fires_df_causes.drop('personal', inplace = True, axis = 1)
all_fires_df_causes.drop('medios', inplace = True, axis = 1)
```

The variable "latlng_explicit" no longer serves any purpose beyond using it to filter out coordinates that ensure the exact location of the fire (rather than a reference like the one used in this dataset, such as the nearest municipality). Similarly, we will also discard the use of the "causa_desc" variable due to the purpose of this exercise, which is the identification of the cause in binary terms. Therefore, maintaining more sophisticated descriptions, in this case, is not meaningful.

```python
all_fires_df_causes.drop('latlng_explicit', inplace = True, axis = 1)
all_fires_df_causes.drop('causa_desc', inplace = True, axis = 1)
```
